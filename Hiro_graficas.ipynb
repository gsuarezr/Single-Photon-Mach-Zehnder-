```python
def Azuma(N,n):
    prob=np.zeros(0)
    prob_2=np.zeros(0)
    A=np.matrix([[ np.cos(n),0],[0,1]])
    jones1= np.matrix([[0],[1]])
    jones0= np.matrix([[1],[0]])
    for i in range(1,N+1):
        alpha =np.pi/(2*i)
        B=np.matrix([[np.cos(alpha),1j*np.sin(alpha)],[1j*np.sin(alpha),np.cos(alpha)]])
        Paola=(B*A)**(i-1)
        prob=np.append(prob,(jones1.T)*Paola*B*jones1)
        prob_2=np.append(prob_2,(jones0.T)*Paola*B*jones1)
        prob_abs=1-(np.absolute(np.square(prob))+np.absolute(np.square(prob_2)))  
    return prob,prob_2,prob_abs
```

# Previously we just repeated to generate the plots of several cases, the function below does a graph for any case wanted for each of the three posible outcomes

```python
def plot_azuma(N,n):
    pd1,pd2,pabs=Azuma(N,n)
    N=np.linspace(1,N,N)
    plt.plot(N,np.square(np.absolute(pd1)),label=r'$PD_{1}$')
    plt.plot(N,np.square(np.absolute(pd2)),label=r'$PD_{2}$')
    plt.plot(N,pabs,label=r'$P_{abs}$')

    plt.xlabel(r'$N$')
    plt.ylabel(r'$ P(N)$')
    plt.xlim(1,50)
    plt.ylim(0,1.05)
    plt.legend()
    return plt.show()
```

# It seems to be not that phase sensitive but let us explore this by doing a plot  for a fixed N let us say fifty, varying with the phase

```python
def Azuma(N,n):
    prob=np.zeros(0)
    prob_2=np.zeros(0)
    jones1= np.matrix([[0],[1]])
    jones0= np.matrix([[1],[0]])
    alpha =np.pi/(2*N)
    B=np.matrix([[np.cos(alpha),1j*np.sin(alpha)],[1j*np.sin(alpha),np.cos(alpha)]])
    for k in np.arange(0,2*np.pi,2*np.pi/100):
        A=np.matrix([[ n*np.exp(1j*k),0],[0,1]])
        Paola=(B*A)**(N-1)
        prob=np.append(prob,(jones1.T)*Paola*B*jones1)
        prob_2=np.append(prob_2,(jones0.T)*Paola*B*jones1)
    prob_abs=1-(np.absolute(np.square(prob))+np.absolute(np.square(prob_2)))  
    return prob,prob_2,prob_abs
```

```python
def plot_azuma(N,n):
    pd1,pd2,pabs=Azuma(N,n)
    k=np.linspace(0,2*np.pi,100)
    plt.plot(k,np.square(np.absolute(pd1)),label=r'$P_{D_{1}}$')
    plt.plot(k,np.square(np.absolute(pd2)),label=r'$P_{D_{2}}$')
    plt.plot(k,pabs,label=r'$P_{abs}$')

    plt.xlabel(r'$\theta$',fontsize=35)
    plt.ylabel(r'$P$',fontsize=35)
    plt.yticks(fontsize=25)
    plt.xticks(np.arange(0,2*np.pi+1, step=np.pi/2),[r'$0$',r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$',r'$2\pi$'],fontsize=25)    
    plt.legend(fontsize=30)
    return plt.show()
```

```python
def hioro(N):
    Z1=[]
    Z2=[]
    Z3=[]
    betaaa=np.arange(0,1.0000000000001,step=N)
    for n in betaaa:
        prob=np.zeros(0)
        prob_2=np.zeros(0)
        A=np.matrix([[ n,0],[0,1]])
        for i in range(1,200):
            alpha =np.pi/(2*i)
            B=np.matrix([[np.cos(alpha),1j*np.sin(alpha)],[1j*np.sin(alpha),np.cos(alpha)]])
            Paola=(B*A)**(i-1)
            prob=np.append(prob,(jones1.T)*Paola*B*jones1)
            prob_2=np.append(prob_2,(jones0.T)*Paola*B*jones1)
        prob_abs=1-(np.absolute(np.square(prob))+np.absolute(np.square(prob_2)))
        Z1.append(prob)
        Z2.append(prob_2)
        Z3.append(prob_abs)
    return Z1,Z2,Z3,betaaa
```

```python
def hioro2(N,intervalo):
    Z1=[]
    Z2=[]
    Z3=[]
    betaaa=np.arange(0,2*np.pi+0.000000001,step=N)
    r=np.arange(0,1+0.0000000001,step=intervalo)
    for beta in r:
        for n in betaaa:
            prob=np.zeros(0)
            prob_2=np.zeros(0)
            A=np.matrix([[ beta*np.exp(1j*n),0],[0,1]])
            for i in range(1,200):
                alpha =np.pi/(2*i)
                B=np.matrix([[np.cos(alpha),1j*np.sin(alpha)],[1j*np.sin(alpha),np.cos(alpha)]])
                Paola=(B*A)**(i-1)
                prob=np.append(prob,(jones1.T)*Paola*B*jones1)
                prob_2=np.append(prob_2,(jones0.T)*Paola*B*jones1)
                prob_abs=1-(np.absolute(np.square(prob))+np.absolute(np.square(prob_2)))
            Z1.append(prob)
            Z2.append(prob_2)
            Z3.append(prob_abs)
    return Z1,Z2,Z3,betaaa
```
