```python
def Azuma(N,n):
    prob=np.zeros(0)
    prob_2=np.zeros(0)
    A=np.matrix([[ np.sqrt(n),0],[0,1]])
    jones1= np.matrix([[0],[1]])
    jones0= np.matrix([[1],[0]])
    for i in range(1,N+1):
        alpha =np.pi/(2*i)
        B=np.matrix([[np.cos(alpha),1j*np.sin(alpha)],[1j*np.sin(alpha),np.cos(alpha)]])
        Paola=(B*A)**(i-1)
        prob=np.append(prob,(jones1.T)*Paola*B*jones1)
        prob_2=np.append(prob_2,(jones0.T)*Paola*B*jones1)
        prob_abs=1-(np.absolute(np.square(prob))+np.absolute(np.square(prob_2)))  
    return prob,prob_2,prob_abs
```

# Previously we just repeated to generate the plots of several cases, the function below does a graph for any case wanted for each of the three posible outcomes

```python
def plot_azuma(N,n):
    pd1,pd2,pabs=Azuma(N,n)
    N=np.linspace(1,N,N)
    plt.plot(N,np.square(np.absolute(pd1)),label=r'$PD_{1}$')
    plt.plot(N,np.square(np.absolute(pd2)),label=r'$PD_{2}$')
    plt.plot(N,pabs,label=r'$P_{abs}$')

    plt.xlabel(r'$N$')
    plt.ylabel(r'$ P(N)$')
    plt.xlim(1,50)
    plt.ylim(0,1.05)
    plt.legend()
    return plt.show()
```

```python
plot_azuma(50,0.6)
```
